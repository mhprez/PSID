```python
# We must now extract the maximum streak of working years.

Streak_Hunt = PSID[Relationships].isin(Valid) # DF of Boolean nature indicating home leadership

Streak_Hunt['Racha'] = '' # Creating an empty column to deposit streak data.

Streak_Hunt[Relationships] = Streak_Hunt[Relationships].replace({True:'T', False:'0'}) # Converting bool to other

A = Streak_Hunt.join(PSID, lsuffix='_Streak', rsuffix='')

def rachita(row, x):
    if row[x] in [1,10]:
        if row[dicc[x][0]] == 0.0:
            row[x+'_Streak'] = '0'
    elif row[x] in [2, 20, 22]:
        if row[dicc[x][1]] == 0.0:
            row[x+'_Streak'] = '0'
    return row[x+'_Streak']

for x in Relationships:
    Streak_Hunt[x] = A.apply(lambda row: rachita(row, x), axis=1)
```

```python
Streak_Hunt['Racha'] = Streak_Hunt.sum(axis=1) # Depositing dirty streak (long string) into column.

Streak_Hunt['Rachab'] = Streak_Hunt['Racha'].str.findall(r'(\D+)') # Separating the various streaks into elements in list.

def get_max_str(lst): # Function to obtain the largest streak out of the list of streaks.
    return max(lst, key=len) # https://blog.finxter.com/how-to-find-the-longest-string-in-a-python-list/

Streak_Hunt['Longest_Spell'] = Streak_Hunt['Rachab'].map(get_max_str) # Obtains the largest string from the list
Streak_Hunt['Length_Spell'] = Streak_Hunt['Longest_Spell'].str.len() # Obtains the length og the largest string.


def funcioncita(row):
    return row['Racha'].index(row['Longest_Spell'])
Streak_Hunt['Beggining_Streak'] = Streak_Hunt.apply(lambda row: funcioncita(row), axis=1)

def otrafuncion(row):
    return Relationships[row['Beggining_Streak']]
Streak_Hunt['Beggining_Streak'] = Streak_Hunt.apply(lambda row: otrafuncion(row), axis=1)
```

```python
A = Streak_Hunt.join(PSID, lsuffix='_Streak', rsuffix='')

def zombor(row):
    start = Relationships.index(row['Beggining_Streak'])
    keys = Relationships[start:start+row['Length_Spell']]
    sal = []
    for col in keys:
        if row[col] in [1,10]:
            sal.append(row[dicc[col][0]])
        elif row[col] in [2,20,22]:
            sal.append(row[dicc[col][1]])
        else:
            sal.append("ERROR")
    return sal
            
    
Streak_Hunt['Salaries'] = A.apply(lambda row: zombor(row), axis=1)
```
